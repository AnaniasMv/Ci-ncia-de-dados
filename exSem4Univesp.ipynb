## Usando uma correlação de Pearson, verifique no database da Marvel, quais são as variáveis que têm maior correlação positiva.



## aqui só aceita dados numericos, entao vamos selecionar apenas os atributos numéricos

#### Filtrando apenas as colunas numéricas

```python
numerical_df = dt_movies.select_dtypes(include=[np.number])
```

#### Calculando a matriz de correlação de Pearson

```python
correlation_matrix = numerical_df.corr(method='pearson')
```

#### Criando uma máscara para a parte superior da matriz de correlação

```python
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
```

## Criando um grafico pra visualização

```python
plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix.corr(), mask=mask, square = True, annot=True, vmin=-1, vmax=1)
plt.show()
```

## Com os dados de Empreendimentos de Geração, indique qual é o atributo que mais tem valores nulos e elimine-o.

```python
valores_nulos = df_relacao.isnull().sum()
valores_nulos
```

```python
atributo_mais_nulos = valores_nulos.idxmax()
atributo_mais_nulos
```

```python
quantidade_nulos = valores_nulos.max()
quantidade_nulos
```

```python
print(f"O atributo com mais valores nulos é '{atributo_mais_nulos}' com {quantidade_nulos} valores nulos.")
```

### removendo o atributo com mais nulos

```python
atributos_mesmos_nulos = valores_nulos[valores_nulos == quantidade_nulos].index.tolist()
atributos_mesmos_nulos
```

```python
df_relacao_sem_nulos = df_relacao.drop(columns=atributos_mesmos_nulos)
print(f"Os atributos com {quantidade_nulos} valores nulos foram removidos: {', '.join(atributos_mesmos_nulos)}")
```

# Com os dados de Empreendimentos de Geração, indique qual é o atributo que tem valores nulos na menor quantidade e elimine as linhas que têm esses atributos com valor nulo.



```python
df_relacao_sem_nulos.isnull().sum()
```

```python
atributo_menos_nulos = valores_nulos[valores_nulos > 0].idxmin()
atributo_menos_nulos
```

```python
df_relacao_sem_nulos = df_relacao_sem_nulos.drop(columns=atributo_menos_nulos)
print(f"O atributo {atributo_menos_nulos} foi removido")
```

#  Com os dados de Empreendimentos de Geração, crie uma nova coluna chamada “mdaPotenciaOutorgadaMW”, baseada na coluna “mdaPotenciaOutorgadakW”. Observe que a nova coluna é em Megawatt e a anterior Kilowatt.

### Converter kW para MW (1 MW = 1000 kW)

```python
df_relacao_sem_nulos
```

```python
df_relacao_sem_nulos['MdaPotenciaInstaladaKW'].unique()
```

```python
df_relacao_sem_nulos['MdaPotenciaInstaladaKW'] = df_relacao_sem_nulos['MdaPotenciaInstaladaKW'].str.replace(',', '.').astype(float)
```

```python
df_relacao_sem_nulos['MdaPotenciaInstaladaKW'].unique()
```

```python
df_relacao_sem_nulos['MdaPotenciaInstaladaMW'] = df_relacao_sem_nulos['MdaPotenciaInstaladaKW'] / 1000
```
